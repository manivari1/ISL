```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
college = pd.read_csv('College.csv')
college.head()




```

```python
#set the index as  the first col
college = college.set_index(['Unnamed: 0'], append=True, verify_integrity=True)
college.rename_axis([None, 'Name'], inplace=True)
print(college.shape)
college.head()
```

```python
#in python we use describe
college.describe()
```

```python
sns.pairplot(college.iloc[:,1:11])
```

```python
sns.boxplot(x = college['Private'],y = college['Outstate'])
```

```python
college['Elite'] = np.where(college['Top10perc'] > 50,'Yes','No')
college.head()
```

```python
college['Elite'].value_counts()
```

```python
elite_colleges = college[college['Elite'] == 'Yes']
elite_colleges.head()
```

```python
# we need to plot by first dividing some of the features into bins  andt then plotting the same
# the features used are ['Enroll','Phd','Termainal']

college['Enroll'] = pd.cut(college['Enroll'],bins=3,labels = ['Low','Medium','High'])
college['PhD'] = pd.cut(college['PhD'],bins = 5,labels = ['Very low','Low','Medium','High','Very High'])
college['Terminal'] = pd.cut(college['Terminal'],bins = 4,labels = ['Very low','Low','High','Very High'])
```

```python
fig = plt.figure()

plt.subplot(2,2,1)
college['Enroll'].value_counts().plot.bar(title = 'Enroll')

plt.subplot(2,2,2)
college['PhD'].value_counts().plot.bar(title = 'PhD')

plt.subplot(2,2,3)
college['Terminal'].value_counts().plot.bar(title = 'Terminal')

#to add space in between the subplots
fig.subplots_adjust(hspace=1)
```

```python
auto = pd.read_csv('Auto.csv')
print(auto.shape)
auto.head()
```

```python
print(auto.nunique())
print(auto.info())
```

```python
#why is horsepower object, when it should be float or int
print(auto['horsepower'].unique())
# we can see that there is a question mark
auto['horsepower'] = auto['horsepower'].replace('?',np.nan)

auto['horsepower'] = auto['horsepower'].astype(np.int32)
```

```python
auto = auto.dropna()
auto.shape
```

```python
info = auto.describe().T
info['range'] = info['max'] - info['min']
info = info[['mean','range','std']]
info.T
```

```python
info = auto.drop(auto.index[10:85]).describe().T
info['range'] = info['max'] - info['min']
info = info[['mean','range','std']]
info.T
```

```python
# we conclude that range is mostly uneffected by removing of the data points, while there is change in mean and std
```
